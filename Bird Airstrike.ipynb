```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
typeSpecification = {"Aircraft Model": str, "Engine Model": str, "Engine3 Position": str, "Engine1 Position":str} 
bird_df = pd.read_csv('bird.csv', dtype=typeSpecification)
```

```python
bird_df.columns
```

```python
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 100)
```

```python
bird_df.shape
```

```python
bird_df.head(50)
```

```python
# Insert Count Column and fill with 1
bird_df["Strike_"] =1
```

```python
pd.isna(bird_df).sum()
```

### Drop Columns

```python
bird_df = bird_df.drop(['Record ID', 'Aircraft Type','Aircraft Make','Aircraft Model','Aircraft Mass','Engine Make','Engine Model',
                         'Engines','Engine Type','Engine1 Position','Engine2 Position','Engine3 Position',
                         'Engine4 Position'], axis = 1)
```

```python
bird_df
```

### Group By Species

```python
species = bird_df.dropna(axis=0, subset=['Species Name'] )
```

```python
species_grp = species.groupby('Species Name').sum()
species_strike = species_grp['Strike_']
species_strike = pd.DataFrame(species_strike)
species_strike.reset_index(inplace=True)
species_strike.sort_values(by='Strike_', ascending=False, inplace=True)
species_strike
```

```python
species_top = species_strike.head(10)
species_top
```

```python
# change the figure size
plt.figure(figsize=(30,15))

al = sns.barplot(x='Species Name', y='Strike_', data=species_top)
#plt.bar(species_top.index, species_top['Strike_'])
#plt.rc('font', size=25) #controls default text size
plt.rc('axes', titlesize=50) #fontsize of the title
plt.rc('axes', labelsize=50) 
plt.rc('xtick', labelsize=20) #fontsize of the x tick labels
plt.rc('ytick', labelsize=20) #fontsize of the x tick labels
plt.xticks(rotation=75)
plt.xlabel("Top 10 Species")
plt.ylabel("Strike Count")
plt.title("Strikes by Species")
al.bar_label(al.containers[0], size=20)
plt.show()

```

### Group By Flight Phase

```python
flight_phase = pd.DataFrame(bird_df["Flight Phase"].value_counts())
flight_phase.reset_index(inplace=True)
flight_phase.rename(columns={"index":"Flight Phase", 'Flight Phase':'Strike Count'}, inplace=True)
flight_phase
```

```python
plt.figure(figsize=(20,15))
ap = sns.barplot(x='Flight Phase', y='Strike Count', data=flight_phase)
plt.xticks(rotation=75)
plt.rc('axes', titlesize=30) #fontsize of the title
plt.rc('axes', labelsize=30) 
plt.rc('xtick', labelsize=20) #fontsize of the x tick labels
plt.rc('ytick', labelsize=20) #fontsize of the x tick labels
plt.xlabel("Flight Phase")
plt.ylabel("Strike Count")
plt.title("Flight Phase Strikes")
plt.rc('bar')
ap.bar_label(ap.containers[0], size=16)
plt.show()
```

### Group By Height

```python
height = bird_df.dropna(axis=0, subset=['Height'] )
```

```python
height.isna().sum()
```

```python
height_vals = height['Height'].value_counts().head(20)
```

```python
#height_vals.reset_index(inplace=True)
hght_vals = pd.DataFrame(height_vals)
#sns.barplot(x='A', y='B', data=height_vals)
hght_vals.reset_index(inplace=True)
hght_vals.rename(columns={"index":"Height", 'Height':'Strike Count'}, inplace=True)
```

```python
hght_vals
```

```python
plt.figure(figsize=(20,15))
af = sns.barplot(x='Height', y='Strike Count', data=hght_vals)
plt.xticks(rotation=75)
plt.rc('axes', titlesize=40) #fontsize of the title
plt.rc('axes', labelsize=30) 
plt.rc('xtick', labelsize=20) #fontsize of the x tick labels
plt.rc('ytick', labelsize=30) #fontsize of the x tick labels
plt.xlabel("Strike Count")
plt.ylabel("Height")
plt.title("Elevation of Strikes")
plt.rc('bar')
af.bar_label(af.containers[0], size=16)
plt.show()
```
