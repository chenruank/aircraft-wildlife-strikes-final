```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
import time
import warnings
warnings.filterwarnings('ignore')
import math
```

```python
birds = pd.read_csv(r'C:\Users\Katie\Documents\Personal\TTS\bird_data.csv', low_memory=False)
```

```python
birds.head()
```

```python
birds['Operator']= birds['Operator'].replace('UNKNOWN',np.nan)
```

```python
birds['Operator ID']= birds['Operator ID'].replace('UNK',np.nan)
birds['Operator'].value_counts().head(10).plot(kind= 'barh')
plt.title("Birdstrikes Per Airline Operator", fontsize=20)
plt.xlabel("Number of Birdstrikes", fontsize=20)
plt.ylabel("Airline Operators", fontsize=20)
```

```python
birds['Operator'].value_counts().head(10)
```

```python
birds['State']= birds['State'].replace('UNKNOWN',np.nan)
birds['State'].value_counts().head(10).plot(kind= 'barh')
plt.title("Birdstrikes Per State", fontsize=20)
plt.xlabel("Number of Birdstrikes", fontsize=20)
plt.ylabel("States", fontsize=20)
```

```python
birds['Incident Month']= birds['Incident Month'].replace('UNKNOWN',np.nan)
birds['Incident Month'].value_counts().head(10).plot(kind= 'barh')
plt.title("Birdstrikes Per Month", fontsize=20)
plt.xlabel("Number of Birdstrikes", fontsize=20)
plt.ylabel("Months", fontsize=20)
```
